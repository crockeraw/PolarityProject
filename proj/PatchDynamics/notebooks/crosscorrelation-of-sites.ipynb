```python
if len(patch2)%2:
    patch2 = patch2[:-1]
    print("yes")
if len(patch1)%2:
    patch1 = patch1[:-1]
    print("yes")
t = np.arange(len(patch1))*15
```

```python
plt.plot(t, patch1, t, patch2)
```

```python
fourier_01 = np.fft.rfft(patch1)
fourier_02 = np.fft.rfft(patch2)
fourier_01[0:2] = 0
fourier_02[0:2] = 0
highpass01 = np.fft.irfft(fourier_01)
highpass02 = np.fft.irfft(fourier_02)
```

```python
plt.plot(t, highpass01, t, highpass02)
```
