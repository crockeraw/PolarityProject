```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
# from cellpose import plot, utils, io
```

```python
pixelX = 2138
micronX = 212.13
pix_to_um = micronX/pixelX
# nM per unit intensity after background subtraction was calculated using Ap4 and Ap5 concentrations measued by FCS.
background = 100
# Both gave similar conversion factors. Ap4 factor is used. nM per unit intesntiy = 0.11766690610151902
bright_to_conc = 0.118
```

```python
sblen = round(5/pix_to_um)
```

```python
im = Image.open('../figures/Ap4_confocal-2-prebud.tif')
a = np.asarray(im)
plt.imshow(a*bright_to_conc - background, cmap="inferno", clim=(0,1000))
plt.colorbar()
plt.plot(range(10,10+sblen), np.ones(sblen)*140, color="white", linewidth=4)
plt.axis('off')
#plt.savefig("../figures/Ap4-prebud.tif")
```
